```python
import requests
import bs4
from bs4 import BeautifulSoup
import time
import os
import csv
```

```python
source=requests.get('https://in.indeed.com/jobs?q=developer&l=').text
soup=BeautifulSoup(source,'lxml')
jobs = soup.find(class_='result')
job_title=jobs.h2.text.strip()
print(job_title)
company_name=jobs.find('span', class_ ='company').text.strip()
print(company_name)
location=jobs.find('div', class_ ='location accessible-contrast-color-location').text.strip()
print(location)
summary=jobs.find('div',class_='summary').text.strip()
print(summary)
```

```python
pages=[10,20,30,40,50]
with open('E:/accountant.csv','a',encoding='utf-8',newline='') as f_output:
    csv_print=csv.writer(f_output)
    
    file_is_empty=os.stat('E:/accountant.csv').st_size==0
    if file_is_empty:
        csv_print.writerow(['Job_Title','Company','Location','Summary','Salary'])
    
 
    for page in pages:
        source=requests.get('https://in.indeed.com/jobs?q=accountant&start={}'.format(page)).text
        soup=BeautifulSoup(source,'lxml')
    
        for jobs in soup.find_all(class_='result'):
            try:
                job_title=jobs.h2.text.strip()
            except Exception as e:
                job_title=None
            print('Job_title:',job_title)
   
            try:
                company_name=jobs.find('span', class_ ='company').text.strip()
            except Exception as e:
                company_name=None
            print('Company_name:',company_name)
   
   
            try:
                location=jobs.find('div', class_ ='location accessible-contrast-color-location').text.strip()
            except Exception as e:
                location=None
            print('Location:',location)

            try:
                summary=jobs.find('div',class_='summary').text.strip()
            except Exception as e:
                summary=None
            print('Summary:',summary)

            try:
                salary=jobs.find('span' ,class_='salaryText').text.strip()
            except Exception as e:
                salary=None
            print('Salary:',salary)
            csv_print.writerow([job_title,company_name,location,summary,salary])
    
            print('-------------------------------------') 
     
            time.sleep(2)
    
```
